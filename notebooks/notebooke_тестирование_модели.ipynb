```python
import pandas as pd
import joblib
```

```python
# 1️⃣ Загружаем тестовый датасет
test_df = pd.read_csv('/kaggle/input/heart-test/heart_test.csv')
```

```python
# 3️⃣ Удаляем лишние колонки, кодируем Gender, обрабатываем пропуски
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
# Кодируем Gender
le = LabelEncoder()
test_df['Gender'] = le.fit_transform(test_df['Gender'])
```

```python
# Импутируем пропуски
imputer = SimpleImputer(strategy='mean')
test_df = pd.DataFrame(imputer.fit_transform(test_df), columns=test_df.columns)
```

```python
ids = test_df['id']
```

```python
# Удаляем id перед подачей в модель
X_test = test_df.drop(columns=['id'])
```

```python
# 4️⃣ Загружаем обученную модель
model = joblib.load('/kaggle/input/heart-attack-model/heart_attack_model.pkl')
```

```python
# 5️⃣ Генерируем предсказания
predictions = model.predict(X_test).astype(int)
```

```python
# 6️⃣ Формируем submission DataFrame
submission = pd.DataFrame({'id': ids.astype(int), 'prediction': predictions})
```

```python
# 7️⃣ Сохраняем submission
submission.to_csv('submission.csv', index=False)
print('✅ submission.csv успешно создан для загрузки')
```
